```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Define the area function in terms of theta and a
def area(theta, a):
    b1 = 1  # minor base
    b2 = b1 + 2 * a * np.cos(theta)  # major base
    h = a * np.sin(theta)            # height
    
    # Trapezoid area function
    A = 0.5 * (b1 + b2) * h
    # We want to maximize the area, so we return the negative of the area
    return -A

# Range of a values
a_values = np.linspace(0.1, 5, 100)  # we avoid 0 to prevent errors in the sine function

# Lists to store the results
theta_optimal_values = []
area_maxima_values = []

# Calculate the optimal angle for each a value
for a in a_values:
    # Initial point for theta
    theta_initial = np.pi / 4
    
    # Specific area function for the current a value
    area_function = lambda theta: area(theta, a)
    
    # Find the angle that maximizes the area
    result = minimize(area_function, theta_initial, bounds=[(0, np.pi)])
    
    # Get the optimal angle and maximum area
    theta_optimal = result.x[0]
    area_maxima = -result.fun
    
    # Store the results
    theta_optimal_values.append(theta_optimal)
    area_maxima_values.append(area_maxima)

# Convert theta_optimal_values from radians to degrees
theta_optimal_values_deg = np.degrees(theta_optimal_values)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(a_values, theta_optimal_values_deg, label='Optimal angle 풪', color='b')
plt.title('Optimal Angle 풪 as a Function of Side Length a')
plt.xlabel('Side length a')
plt.ylabel('Optimal angle 풪 (degrees)')
plt.legend()
plt.grid(True)
plt.show()
```
